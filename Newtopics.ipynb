```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
import matplotlib.pyplot as plt

# Step 1: Create sample dictionary data and convert to DataFrame
data = {
    "Product": ["A", "B", "A", "C", "B", "C", "A", "B", "C", "A"],
    "Category": ["Electronics", "Furniture", "Electronics", "Toys", "Furniture", "Toys", "Electronics", "Furniture", "Toys", "Electronics"],
    "Region": ["North", "South", "East", "West", "North", "South", "East", "West", "North", "South"],
    "Sales": [250, 300, 150, 200, 175, 225, 275, 325, 125, 180],
    "Profit": [20, 50, 30, 40, 25, 35, 15, 55, 45, 30]
}

df = pd.DataFrame(data)

# Step 2: Encoding
# Label Encoding
le = LabelEncoder()
df["Product_encoded"] = le.fit_transform(df["Product"])
df["Category_encoded"] = le.fit_transform(df["Category"])
df["Region_encoded"] = le.fit_transform(df["Region"])

# One-Hot Encoding
df_one_hot = pd.get_dummies(df, columns=["Product", "Category", "Region"])

# Display DataFrame after encoding
print("Data after Label Encoding and One-Hot Encoding:")
print(df_one_hot)

# Step 3: Standardization
scaler = StandardScaler()
df[["Sales_standardized", "Profit_standardized"]] = scaler.fit_transform(df[["Sales", "Profit"]])

# Display DataFrame after standardization
print("\nData after Standardization:")
print(df[["Sales", "Profit", "Sales_standardized", "Profit_standardized"]])

# Step 4: Visualization
# Pie Chart: Distribution of Product
plt.figure(figsize=(8, 5))
df["Product"].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=["skyblue", "lightgreen", "salmon"])
plt.title("Product Distribution")
plt.ylabel("")  # Hide y-axis label for clarity
plt.show()

# Boxplot: Distribution of Sales
plt.figure(figsize=(8, 5))
plt.boxplot(df["Sales"], vert=False)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.show()

# Bar Graph: Average Profit by Category
plt.figure(figsize=(8, 5))
df.groupby("Category")["Profit"].mean().plot(kind="bar", color="lightcoral")
plt.title("Average Profit by Category")
plt.xlabel("Category")
plt.ylabel("Average Profit")
plt.show()
```
